# **Collecting Data from [PostJobFree.com](https://www.postjobfree.com/)**

## Importing the needed libraries

```python
import requests
from bs4 import BeautifulSoup
import json
import time
from random import seed
from random import randint
```

## Function to Write Out the result to a JSON file

```python
def WriteOutJson(file_path, results):
    print("Writing out data...")
    with open(file_path, "w") as file:
        json.dump(results, file)
    print("Done!")
```

## Setting Up the Parameters for the Collected Data

 **Locations** | **Domains** | **Keywords**
 --- | --- | ---
 *5* | *11* | *65*

```python
locations=['Canada', 'Egypt', 'US', 'UK', 'Australia']
domains={}
domains['Engineering']=['biomedical+engineer', 'industrial+engineer', 'civil+engineer', 'chemical+engineer', 'electrical+engineer', 'computer+engineer', 'petroleum+engineer', 'Mechanical+Engineer']
domains['Entertainment']=['photographer', 'graphic+designer', 'fashion+designer', 'music+producer', 'filmmaker', 'animator']
domains['Administration']=['secretary', 'human+resource+manager', 'administrative+assistant', 'office+administartor']
domains['Business']=['business+consultant', 'business+development+manager', 'accountant', 'cost+estimator', 'Financial+Advisor', 'Audit+clerk', 'product+owner', 'product+manager']
domains['Communication']=['journalist', 'public+relations+agent', 'event+planner', 'social+media+manager', 'Interpreter',  'Translator']
domains['Education']=['school+principal', 'teacher', 'professor', 'special+education+teacher', 'school+Psychologist']
domains['Science']=['physicist', 'microbiologist', 'archaeologist', 'laboratory+technician']
domains['Technology']=['software+engineer', 'data+scientist', 'full+stack+developer', 'data+engineer', 'backend-developer', 'computer+system+analyst', 'database+administrator', 'web+developer']
domains['Medicine']=['nurse', 'veterinarian', 'dental+assistant', 'anesthesiologists', 'Dentist', 'Pharmacist', 'Physical+Therapist+Assistant', 'Epidemiologist', 'paramedic']
domains['Sales']=['sales+associate', 'sales+manager', 'account+executive', 'sales+development+representation']
domains['Marketing']=['Market+Research+Analyst', 'marketing+assistant', 'marketing+manager', 'product+owner', 'product+manager']
results = []
ID=0
```

## Function to Generate the Resumes and Append it with the Parameters to a list (results)

```python
def ResumeGen(domains, locations, results, ID):
    for key, value in domains.items():
        for word in value:
            for location in locations:
                url = 'https://www.postjobfree.com/resumes?q=&n=&t=' + word + '&d=&l=' + location + '&radius=50&r=100'
                response = requests.get(url)
                soup = BeautifulSoup(response.content, 'html.parser')
                title_tags = soup.find_all('h3', attrs={'class': 'itemTitle'})
                links = ["https://www.postjobfree.com" + title_tag.a['href'] for title_tag in title_tags]
                for link in links:
                    res = requests.get(link)
                    print(res.status_code, link)
                    content = BeautifulSoup(res.content, 'html.parser')
                    ID+=1
                    results.append({
                        'ID': ID,
                        'job_title': content.find('div', attrs={'class': 'leftColumn'}).find('h1').get_text(),
                        'resume': content.find('div', attrs={'class': 'normalText'}).get_text()[:-23],
                        'link' : link,
                        'domain': key,
                        'location': location
                    })
            seed(1)
            for _ in range(10):
                x = randint(0, 10)
                time.sleep(x)
```

```python
ResumeGen(domains, locations, results, ID)
```

```python
WriteOutJson('./BulkResumesAll.json', results)
```
